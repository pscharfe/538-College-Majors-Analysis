# Regression analysis for output variables
## Median income and unemployment

```python
import statsmodels.formula.api as smf
from scipy import stats
```

```python
outvars_model = smf.ols("Median ~ Unemployment_rate", data= df_recent).fit()

outvars_model.summary()

# The correlation given in Line 17 was -0.108833
```

```python
medwomen_model = smf.ols("Median ~ ShareWomen", data= df_recent).fit()

medwomen_model.summary()
```

```python
stats.probplot(medwomen_model.resid, plot= plt)
plt.title("Women and Income Regression Residuals Probability Plot")
```

```python
unempwom_model = smf.ols("Unemployment_rate ~ ShareWomen", data= df_recent).fit()

unempwom_model.summary()
```

```python
stats.probplot(unempwom_model.resid, plot= plt)
plt.title("Women and Unemployment Regression Residuals Probability Plot")
```

```python
df_recent['SharePT'] = df_recent['Part_time']/df_recent['Total']

# To see the effect of part-time work on our output variables, we need to determine the *share* of part-time workers
# The raw data only give the absolute number of part-time workers, which is not helpful
```

```python
prod_model = smf.ols("SharePT ~ ShareWomen", data= df_recent).fit()

prod_model.summary()
```

```python
prod_model = smf.ols("Median ~ ShareWomen + SharePT", data= df_recent).fit()

prod_model.summary()
```

```python
df_recent['Interact_PT_Women'] = df_recent['ShareWomen']*df_recent['SharePT']

# If % of women and % of part-time are related, and since both affect income, we add a new composite variable
# This composite variable is the product of both % of women and % of part-time
```

```python
interactmed_model = smf.ols("Median ~ Interact_PT_Women", data= df_recent).fit()

interactmed_model.summary()
```

```python
stats.probplot(interactmed_model.resid, plot= plt)
plt.title("PT-Women Interaction Regression Residuals Probability Plot")
```

```python
prod_model = smf.ols("Median ~ ShareWomen + SharePT + Interact_PT_Women", data= df_recent).fit()

prod_model.summary()

# Despite both being related, the composite variable of women and part-time is not significant
```

```python
df_recent = pd.read_csv("recent-grads.csv") #Reading the Whole Sale Customer dataset again in a 
# dataframe using Pandas

#Make a copy of the dataframe before we start a transformation
df_recent_new = df_recent

#Standardize
clmns = ['Median','Unemployment_rate','ShareWomen', 'SharePT']
Recent_Clusters = stats.zscore(df_recent_new[clmns])
```
